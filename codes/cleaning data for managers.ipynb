```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df['Appointed'] = pd.to_datetime(df['Appointed'], format='%d-%b-%y', errors='coerce')
```

```python
df['Salary (million)'] = df['Salary'].str.replace(' million', '').astype(float)
df.drop(columns=['Salary'], inplace=True)
```
