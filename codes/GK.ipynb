```python
df = pd.read_csv("goalkeepers_2.csv")

print(df.info())
print(df.head())
```

```python
missing = df.isnull().sum()
print(missing[missing > 0]) 
```

```python
cols_to_drop = ['SoT%', 'G/Sh', 'G/SoT', 'Dist', 'npxG/Sh','Tkl%','Succ%','Tkld%']  
df.drop(columns=cols_to_drop, inplace=True)
```

```python
df
```

```python
cols_to_fill = [
    'Won%', 'Save%', 'CS%', 'PSxG/SoT', 'Cmp%_stats_keeper_adv','AvgDist']

df[cols_to_fill] = df[cols_to_fill].fillna(0)

print(df[cols_to_fill].isnull().sum())
```

```python
missing = df.isnull().sum()
print(missing[missing > 0]) 
```

```python
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

```python
# This keeps only columns that have at least one non-zero value.
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
df
```

```python
df = df.drop(columns=["Rk"])

df["gk_id"] = range(1, len(df) + 1)

df = df[["gk_id"] + [col for col in df.columns if col != "gk_id"]]
```

```python
df.to_csv("cleaned_GK_2.csv", index=False)
print("Cleaned CSV saved as 'cleaned_GK.csv'")
```
